FIND THE TOTAL PRICE

```python
df['Totalcost']=((df['Quantity']*df['UnitPrice'])+df['ShippingCost'])-df['Discount']
df.info()
```

1. WHAT IS THE AVERAGE SHIPPING COST?

```python
df["ShippingCost"].mean()
```

2. WHICH PRODUCT HAS THE HIGH MARKET DEMAND?

```python
product=df.groupby("Description")["Quantity"].sum()

print(product)

product.plot(kind='bar',figsize=(10,6),color='skyblue',title='market demand',
ylabel='quantity sold',xlabel='products')
plt.show()
```

 3. GRAPHICALLY IDENTIFY THE RETURN STATUS?

```python

df["ReturnStatus"].max()
total=df.groupby("ReturnStatus")["Totalcost"].sum()
plt.pie(total, labels=total.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.show()
```

4. WHICH TYPE OF SALESCHANNEL HAS MOSTLY USED?

```python
# df["SalesChannel"].max()
channel_sales = df.groupby('SalesChannel')['Totalcost'].sum()
plt.pie(channel_sales, labels=channel_sales.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.show()
```

5  WHICH PAYMENT METHOD IS MOSTLY PREFFERD?

```python
a=df["PaymentMethod"].value_counts()
print(a)
a.plot(kind='pie',autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.show()
```

6 CALCULATE THE NO OF SHIPMENTPSERVICE DONE BY UPS?

```python
shipmentservice=df[df["ShipmentProvider"]=="UPS"]["ShipmentProvider"].count()
print(shipmentservice)
```

7 TO IDENTIFY WHICH WAREHOUSE IS PERFORMING THE BEST?

```python
warehouse_sales=df.groupby("WarehouseLocation")["Totalcost"].sum().sort_values(ascending=False)
print(warehouse_sales)
warehouse_sales.plot(kind='bar',figsize=(10,6),color='green',title='market demand',
ylabel='quantity sold',xlabel='products')
plt.show()
```

8 HOW MANY SALES IN A YEAR 2020?

```python
df["InvoiceDate"]=pd.to_datetime(df["InvoiceDate"])
startdate='2020-01-01'
enddate='2020-12-31'
new=(df["InvoiceDate"]>=startdate)&(df["InvoiceDate"]<=enddate)
print(new.sum())
```

10 WHAT ARE THE TOP 5 MOST POPULAR PRODUCTS?

```python
popular_products=df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(5)
print(popular_products)
popular_products.plot(kind='bar',figsize=(10,6),color='red',title='Top 5 most popularproducts',
ylabel='quantity sold',xlabel='products')
plt.show()
```

11 WHAT IS THE TOTAL SALES PER MONTH?

```python
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])
monthly_sales = df.groupby(df['InvoiceDate'].dt.to_period('M'))['Totalcost'].sum()
print(monthly_sales)
```

```python

monthly_sales = df.groupby(df['InvoiceDate'].dt.to_period('M'))['Totalcost'].sum()
plt.figure(figsize=(10, 6))  
monthly_sales.plot(kind='line', marker='o', linestyle='-', color='b', title='Monthly Revenue')
plt.xlabel('Month')
plt.ylabel('Totalcost')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()m
```

12 COMPARE WAREHOUSE PERFORMANCE ON A MONTHLY BASIS?

```python
performance= df.groupby([df['InvoiceDate'].dt.to_period('M'),'WarehouseLocation'])['Quantity'].sum()
monthly_performance=performance.unstack().fillna(0)
monthly_performance.plot(kind='line',figsize=(12,6),title='performance_overtime',ylabel='quantitysold',xlabel='month')

plt.show()

```
